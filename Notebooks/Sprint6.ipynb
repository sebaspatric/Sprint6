```python
import numpy as np

# Definir puntos y centroides iniciales
points = np.array([[10, 4, 50],
                   [8, 99, 31],
                   [34, 44, 86],
                   [9, 50, 57],
                   [46, 77, 75],
                   [68, 30, 14],
                   [80, 25, 40],
                   [50, 35, 60]])

initial_centroids = np.array([[47.5, 37.5, 21.8],
                              [53.2, 22.4, 75.3]])

# Función para calcular distancia euclidiana
def euclidean_distance(point, centroid):
    return np.sqrt(np.sum((point - centroid) ** 2))

# Función para asignar puntos a grupos
def assign_points_to_clusters(points, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for point in points:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        closest_centroid_idx = np.argmin(distances)
        clusters[closest_centroid_idx].append(point)
    return clusters

# Función para actualizar centroides
def update_centroids(clusters):
    return [np.mean(cluster, axis=0) for cluster in clusters]

# Iteraciones del algoritmo K-means
for i in range(3):
    # Asignar puntos a grupos
    clusters = assign_points_to_clusters(points, initial_centroids)
    print(f"Iteración {i + 1}:")
    for j, cluster in enumerate(clusters):
        print(f"Grupo {j + 1}: {cluster}")
    # Actualizar centroides
    new_centroids = update_centroids(clusters)
    print(f"Nuevos centroides: {new_centroids}")
    print()
    # Verificar convergencia
    if np.array_equal(new_centroids, initial_centroids):
        print("Convergencia alcanzada.")
        break
    else:
        initial_centroids = new_centroids
```

## 5. Enuncie al menos tres diferencias entre el análisis factorial y el método de componentes principales.


El análisis factorial y el método de componentes principales (PCA) son técnicas utilizadas en el análisis de datos multivariados para reducir la dimensionalidad y extraer información importante de conjuntos de datos complejos. Aquí hay tres diferencias clave entre el análisis factorial y el método de componentes principales:

1. **Objetivo y enfoque:**
   - **Análisis factorial:** El análisis factorial busca identificar las relaciones subyacentes entre un conjunto de variables observadas y explicar estas relaciones en términos de un número menor de factores latentes no observados. Se utiliza para comprender la estructura subyacente de los datos y revelar las dimensiones subyacentes que explican la variabilidad común entre las variables.
   - **Componentes principales:** El PCA busca reducir la dimensionalidad de un conjunto de datos manteniendo la mayor cantidad posible de variabilidad original. Se basa en encontrar nuevas variables (llamadas componentes principales) que son combinaciones lineales de las variables originales y que capturan la mayor parte de la varianza en los datos. El objetivo principal es reducir la complejidad del conjunto de datos mientras se retiene la mayor cantidad posible de información.

2. **Suposiciones subyacentes:**
   - **Análisis factorial:** El análisis factorial asume que las variables observadas son medidas imperfectas o indicadores de factores subyacentes no observados. Estos factores son constructos hipotéticos que no se miden directamente pero que se supone que influyen en las observaciones.
   - **Componentes principales:** El PCA no implica suposiciones sobre las relaciones causales o subyacentes entre las variables observadas. Se basa únicamente en la estructura de covarianza (o correlación) entre las variables y busca encontrar una representación de baja dimensionalidad que mantenga la mayor cantidad de varianza original.

3. **Interpretación de los resultados:**
   - **Análisis factorial:** Los factores extraídos en el análisis factorial son constructos latentes que representan relaciones subyacentes entre las variables observadas. La interpretación de los factores implica comprender qué aspectos del fenómeno están representados por cada factor y cómo se relacionan entre sí.
   - **Componentes principales:** Los componentes principales en PCA son combinaciones lineales de las variables originales que capturan la mayor parte de la varianza en los datos. La interpretación de los componentes principales implica comprender qué variables contribuyen más a cada componente y cómo se relacionan entre sí en términos de covarianza o correlación.

En resumen, aunque tanto el análisis factorial como el PCA son técnicas de reducción de dimensionalidad, difieren en sus objetivos, suposiciones subyacentes y enfoques de interpretación de resultados.

## 6. ¿Qué significa que el método de clusterización sea jerárquico?

En el contexto de la clusterización o agrupamiento (clustering) de datos, un método jerárquico implica que los grupos o clústeres se organizan en una estructura jerárquica, es decir, en una serie de niveles o niveles de agrupamiento. Esto significa que los datos pueden ser agrupados en múltiples niveles, donde los clústeres en un nivel superior pueden contener clústeres más pequeños en niveles inferiores.

Un método de clustering jerárquico puede ser aglomerativo o divisivo:

1. **Aglomerativo (Bottom-up):** En este enfoque, cada punto comienza en su propio clúster y los clústeres se van fusionando gradualmente en clústeres más grandes a medida que avanzamos en la jerarquía. En cada paso, los dos clústeres más similares se fusionan en uno solo, y este proceso continúa hasta que todos los puntos están en un solo clúster.

2. **Divisivo (Top-down):** En este enfoque, todos los puntos comienzan en un solo clúster y este clúster se divide gradualmente en clústeres más pequeños a medida que avanzamos en la jerarquía. En cada paso, un clúster se divide en dos clústeres más pequeños basados en algún criterio de división, y este proceso continúa hasta que cada punto está en su propio clúster.

En ambos enfoques, la jerarquía resultante se puede visualizar como un dendrograma, que es una representación gráfica de la estructura jerárquica de los clústeres. Cada nivel en el dendrograma representa un nivel de agrupamiento y las ramas del dendrograma muestran cómo los clústeres se fusionan o dividen a lo largo del proceso. La elección entre un enfoque aglomerativo o divisivo depende del conjunto de datos y del objetivo específico del análisis.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

def elbow_method(data, max_clusters=8):
    inertia = []
    k_range = range(1, max_clusters + 1)
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
    return k_range, inertia

def silhouette_method(data, max_clusters=8):
    silhouette_scores = []
    k_range = range(2, max_clusters + 1)
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(data)
        silhouette_avg = silhouette_score(data, cluster_labels)
        silhouette_scores.append(silhouette_avg)
    return k_range, silhouette_scores

def plot_elbow_method(k_range, inertia):
    plt.plot(k_range, inertia, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')
    plt.show()

def plot_silhouette_method(k_range, silhouette_scores):
    plt.plot(k_range, silhouette_scores, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Method')
    plt.show()

# Uso de las funciones
k_range_elbow, inertia = elbow_method(data_pca_df)
plot_elbow_method(k_range_elbow, inertia)

k_range_silhouette, silhouette_scores = silhouette_method(data_pca_df)
plot_silhouette_method(k_range_silhouette, silhouette_scores)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

def find_optimal_number_of_clusters(data, max_clusters=8):
    # Método del codo
    inertia_values = []
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        inertia_values.append(kmeans.inertia_)

    # Método de la silueta
    silhouette_scores = []
    for k in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(data)
        silhouette_avg = silhouette_score(data, cluster_labels)
        silhouette_scores.append(silhouette_avg)

    # Encontrar el número óptimo de clusters usando el método del codo
    optimal_k_elbow = inertia_values.index(min(inertia_values)) + 1

    # Encontrar el número óptimo de clusters usando el método de la silueta
    optimal_k_silhouette = silhouette_scores.index(max(silhouette_scores)) + 2

    return optimal_k_elbow, optimal_k_silhouette

# Uso de la función para encontrar el número óptimo de clusters
optimal_k_elbow, optimal_k_silhouette = find_optimal_number_of_clusters(data_pca_df)
print("Número óptimo de clusters según el método del codo:", optimal_k_elbow)
print("Número óptimo de clusters según el método de la silueta:", optimal_k_silhouette)
```

## Descripción del Problema
El área de marketing de una empresa desea generar una clusterización para poder tener mayor éxito con sus campañas. Se proporcionó una base de datos que contiene información sobre clientes, incluyendo características demográficas y comportamiento de compra.

1. **Carga de Datos:** Se cargaron los datos desde el archivo CSV "valuacion_mkt_campaign.csv".
2. **Selección de Características:** Se seleccionaron las columnas relevantes para el análisis, incluyendo la edad, antigüedad, ingreso, recencia de compra y gastos en diferentes categorías de productos.
3. **Estandarización de Datos:** Los datos fueron estandarizados para asegurar que todas las características tuvieran la misma escala y contribución al análisis.

## Análisis de Clústeres

- Se identificaron y descartaron registros anómalos correspondientes al 5% de la muestra, utilizando las columnas de comportamiento de compra.

### Feature Extraction
- Se redujo el número de variables utilizando un método de extracción de características, buscando una representación que explicara el 90% de la varianza original de los datos.

### Clustering
- Se utilizó el algoritmo K-Means para clusterizar el comportamiento de los clientes.
- Se determinó el número óptimo de clústeres que no excediera los 8, considerando las restricciones de la empresa de marketing.

## Resultados
- Se generaron clústeres que agrupan a los clientes según sus características demográficas y comportamiento de compra.
- Se identificaron patrones claros de comportamiento dentro de cada clúster, lo que proporciona información valiosa para personalizar las estrategias de marketing.
- Se calcularon y graficaron los centroides de cada clúster para visualizar la ubicación media de los clientes en el espacio de características.

## Conclusiones
El análisis de clústeres permitió identificar segmentos de clientes con características similares, lo que puede ayudar a la empresa de marketing a personalizar sus estrategias y campañas para satisfacer las necesidades específicas de cada grupo. Estos hallazgos proporcionan información valiosa para mejorar la efectividad y el rendimiento de las campañas de marketing.
